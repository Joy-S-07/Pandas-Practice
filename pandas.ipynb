### Editing Methods in ```.csv``` file
1. ```dropna()``` function : To avoid rows which does not have any information like the email or DOB
* Syntax : ```dropna()```
2. ```fillna(data)``` function: To add data which will be filled in blank spaces <br>
* Syntax: ```fillna(N.A. , inplace = True)``` --> Here ```inplace = True``` is not recommended as it will make changes in Original Data Set or in ```.csv``` file
3. ```rename()``` function : Used to rename any column name. Here also ```inplace = True``` is used but do not use unconsciously otherwise Original Data will be changed


````python
people = pd.read_csv('./data/people-100.csv')
people.head()
````

````python
people[people["Sex"]=="Male"]["Email"]
````

````python
# Here the NaN are the blank Spaces.
````

````python
people.dropna()
````

````python
# Here the 5th index is missing now for dropna() function
````

````python
people.fillna("Missing")
````

````python
# Now the 5th Row Last Name Column is rewritten Missing but not changed the Original file. To change the Original File Use inplace = True in the function
````

````python
people.rename(columns={"User Id" : "ID"})
````

### ```len()``` function and adding column using pandas (Manually &  using Custom functions)


````python
len(people)
````

````python
people["manual_column"] = [0 for i in range(len(people))]
people
````

````python
people["Index"] = people["Index"].astype(int)
def func(a):
    return a*a

people["function_column"] = people["Index"].apply(func)
people
````

### Concatenating DataFrames : When Want to make to DataFrames into One

````python
dataFrame_1 = pd.DataFrame({"name":["Gojo" , "Geto" , "Shoko"], "marks":[100 , 90 , 80]})
dataFrame_2 = pd.DataFrame({"name":["Tanjiro" , "Zenitsu" , "Inosuke"] , "marks":[100 , 90 , 80]})

dataFrame_1
````

````python
dataFrame_2
````

````python
concat_dataFrame = pd.concat([dataFrame_1, dataFrame_2])
concat_dataFrame
````

### Merging DataFrames : Used to merger two data frames to get common data based on particular column

````python
dataFrame_3 = pd.DataFrame({"name":["Gojo" , "Geto" , "Shoko" , "Sukuna"], "marks":[100 , 90 , 80 , 99]})
dataFrame_4 = pd.DataFrame({"name":["Gojo" , "Geto" , "Shoko"], "roll":[10 , 11 , 1]})
````

````python
dataFrame_3
````

````python
merged = pd.merge(dataFrame_3 , dataFrame_4 , on="name")
merged
````
